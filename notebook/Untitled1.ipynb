```python
import pandas as pd
import json

df = pd.read_parquet("../data/history.parquet")
with open("../data/dictionary.json") as fp:
    dictionary = json.load(fp)

keys = [elem for elem in dictionary]
values = [dictionary[idx] for idx in keys]
df_dict = pd.DataFrame({"document" : keys, "pos" : values})

df = df.merge(df_dict, on="document")
```

```python
df = df.sort_values("id")
df
```

```python
sample = df.iloc[0:10000]

def top_n(frame, n, columns):
    length = len(frame)
    frame = frame.sort_values(by=columns)
    if length >= n:
        return frame[-n:]
    else:
        return frame.append([frame.iloc[-1]] * (n - length))

sample = sample.groupby("id", group_keys=False).apply(top_n, n=15, columns=["timestamp", "document"])
```

```python
df = df.groupby("id", group_keys=False).apply(top_n, n=15, columns=["timestamp", "document"])
```
